```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe_track = pd.read_csv("tracks.csv")
dataframe_track
```

```python
dataframe_track.info()
```

```python
dataframe_track.columns
```

```python
# Most popular song
most_popular_track = dataframe_track.sort_values('popularity',ascending=False)
most_popular_track.head(10)
```

```python
# Least popular song
least_track = dataframe_track.sort_values('popularity',ascending=True)
least_track.head(10)
```

```python
# Stastics view of the data
dataframe_track.describe().transpose()
```

```python
dataframe_track.set_index('release_date',inplace=True)
dataframe_track.index=pd.to_datetime(dataframe_track.index)
dataframe_track.head(5)
```

```python
# Change unit of the duration from ms to s
dataframe_track['duration'] = dataframe_track['duration_ms'].apply(lambda x: round(x/1000))
dataframe_track.drop('duration_ms',inplace=True,axis=1)
```

```python
dataframe_track.duration.head()
```

```python
# Remove not necessary columns and find Pearson correlation and plot graph
correlation_dataframe = dataframe_track.drop(['key','mode','explicit'],axis = 1).corr(method='pearson')
plt.figure(figsize=(15,7))
heatmap = sns.heatmap(correlation_dataframe,annot=True,vmin= -1,vmax = 1 ,center = 0,cmap='coolwarm',linewidth=1,linecolor = 'black')
heatmap.set_title("Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
#find kendall correlation and plot graph
correlation_dataframe = dataframe_track.drop(['key','mode','explicit'],axis = 1).corr(method='kendall')
plt.figure(figsize=(15,7))
heatmap = sns.heatmap(correlation_dataframe,annot=True,vmin= -1,vmax = 1 ,center = 0,cmap='coolwarm',linewidth=1,linecolor = 'black')
heatmap.set_title("Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
#find spearman correlation and plot graph
correlation_dataframe = dataframe_track.drop(['key','mode','explicit'],axis = 1).corr(method='spearman')
plt.figure(figsize=(15,7))
heatmap = sns.heatmap(correlation_dataframe,annot=True,vmin= -1,vmax = 1 ,center = 0,cmap='coolwarm',linewidth=1,linecolor = 'black')
heatmap.set_title("Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Below results are same for all correlation methods
# We know the highest correlation between energy and looudness(0.8) and medium correlation between valence and energy (0.5) from heatmap
# Lowest correlation between acousticness and energy (-0.7)
```

```python
# first take sample data from whole data
sample_dataframe = dataframe_track.sample(int(0.004 * len(dataframe_track)))
print(len(sample_dataframe))
```

```python
# We plot the regression line to find exact relation between two variable (loudness vs. energy)
plt.figure(figsize=(10,6))
sns.regplot(data= sample_dataframe,y = 'loudness',x='energy',color='c',line_kws={'color':'red'}).set(title='Loudness vs Energy correlation')
```

```python
# Popularity and acousticness are oppesite relation 
plt.figure(figsize=(10,6))
sns.regplot(data= sample_dataframe,y = 'popularity',x='acousticness',color='r',line_kws={'color':'blue'}).set(title='Popularity vs acousticness correlation')
```

```python
# Add new column name year come from relase date

dataframe_track['dates'] = dataframe_track.index.get_level_values('release_date')
dataframe_track.dates = pd.to_datetime(dataframe_track.dates)
years = dataframe_track.dates.dt.year 
```

```python
# Plot the graph indicate how songs per year are released

sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of song per year')
```

```python
total_duration = dataframe_track.duration
fig, ax = plt.subplots(figsize = (18,7))
fig = sns.barplot(x=years,y = total_duration, ax =ax,errwidth=False).set(title = "Years vs Duration")
plt.xticks(rotation = 90)
```

```python
# We analye the how duration of song increase and decrease based on the year 

total_duration = dataframe_track.duration
sns.set_style(style='whitegrid')
fig, ax = plt.subplots(figsize = (18,7))
fig = sns.lineplot(x=years,y = total_duration, ax =ax).set(title = "Years vs Duration")
```

```python
dataframe_genre = pd.read_csv("SpotifyFeatures.csv")
dataframe_genre.head()
```

```python
# Change unit of the duration from ms to s
dataframe_genre['duration'] = dataframe_genre['duration_ms'].apply(lambda x: round(x/1000))
dataframe_genre.drop('duration_ms',inplace=True,axis=1)
```

```python
# How many seconds in different different genre through graph
plt.figure(figsize=(15,8))
plt.title("Duration of the songs in Different genre")
sns.color_palette('rocket',as_cmap=True)
sns.barplot(y='genre',x='duration',data = dataframe_genre)
plt.xlabel('Duration in seconds')
plt.ylabel('Genres')
```

```python
# top genres based on popularity
sns.set_style(style='darkgrid')
plt.figure(figsize=(15,8))
popular_genre = dataframe_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(y ='genre',x= 'popularity',data=popular_genre).set(title = 'Top 5 Genres by popularity')
```

```python
dataframe_genre.columns
```

```python
dataframe_genre.genre.unique()
```

```python
#popular genre based on artist
popular_genre_artist = dataframe_genre.groupby('genre')[('artist_name')].count()
popular_genre_artist.sort_values(ascending=False)
```
